```python
class Inventory:
    def __init__(self):
        self.inventory = {}

    def add_item(self, sku, item_name, stock_count, price):
        self.inventory[sku] = {'item_name': item_name, 'stock_count': stock_count, 'price': price}

    def update_item(self, sku, item_name=None, stock_count=None, price=None):
        if sku in self.inventory:
            if item_name:
                self.inventory[sku]['item_name'] = item_name
            if stock_count:
                self.inventory[sku]['stock_count'] = stock_count
            if price:
                self.inventory[sku]['price'] = price

    def check_item_details(self, sku):
        return self.inventory.get(sku, 'Item not found')

    def check_stock_levels(self):
        for sku, details in self.inventory.items():
            if details['stock_count'] <= 5:
                print(f"Low stock alert for {details['item_name']}. Only {details['stock_count']} left.")
```

```python
    def place_order(self, sku, quantity):
        if self.inventory[sku]['stock_count'] <= 5:
            print(f"Placing order for {quantity} of {self.inventory[sku]['item_name']}")
            self.inventory[sku]['stock_count'] += quantity
```

LSTM approach

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
from sklearn.metrics import mean_squared_error
import math

# Load your data
#df = pd.read_csv('your_data.csv')

# Keep only relevant columns for demand forecasting
df = df[['Product type', 'SKU', 'Price', 'Availability', 'Number of products sold', 'Order Date']]
#columns_to_drop =['Revenue generated','Customer demographics','Stock levels','Lead times','Order quantities','Shipping times','Shipping carriers','Shipping costs','Supplier name','Location','Lead time','Production volumes','Manufacturing lead time','Manufacturing costs','Inspection results','Defect rates','Transportation modes','Routes','Costs','Delivery time','Delivery times','Total Revenue','Total Cost','Delivery Date']
#df = df.drop(columns=columns_to_drop)
#df.head()

# Convert 'Order Date' column to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Sort data by date
df = df.sort_values('Order Date')



# First, normalize the dataset
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(df['Number of products sold'].values.reshape(-1,1))

# Split the dataset into train and test sets
train_size = int(len(dataset) * 0.80)
test_size = len(dataset) - train_size
train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]

# Convert an array of values into a dataset matrix
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i + look_back, 0])
    return np.array(dataX), np.array(dataY)

# Reshape into X=t and Y=t+1
look_back = 1
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)

# Reshape input to be [samples, time steps, features]
trainX = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
testX = np.reshape(testX, (testX.shape[0], 1, testX.shape[1]))

# Create and fit the LSTM network
model = Sequential()
model.add(LSTM(4, input_shape=(1, look_back)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(trainX, trainY, epochs=100, batch_size=1, verbose=2)

# Make predictions
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)

# Invert predictions
trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainY])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testY])

# Calculate root mean squared error
trainScore = math.sqrt(mean_squared_error(trainY[0], trainPredict[:,0]))
print('Train Score: %.2f RMSE' % (trainScore))
testScore = math.sqrt(mean_squared_error(testY[0], testPredict[:,0]))
print('Test Score: %.2f RMSE' % (testScore))
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

series = df.set_index('Order Date')['Number of products sold']

# Split the data into training/testing sets
train_size = int(len(series) * 0.80)
train, test = series[0:train_size], series[train_size:len(series)]

# Fit a SARIMAX(1, 1, 1)x(1, 1, 1, 12) model
model = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
model_fit = model.fit(disp=False)

# Make predictions
predictions = model_fit.predict(start=len(train), end=len(train) + len(test) - 1)

# Calculate the RMSE
mse = mean_squared_error(test, predictions)
rmse = math.sqrt(mse)
print('RMSE: %.2f' % rmse)
```

ABC Analysis

```python

# Calculate total value, defect rate, and revenue for each product type
total_value = df.groupby('Product type')['Order quantities'].sum()
average_defect_rate = df.groupby('Product type')['Defect rates'].mean()
total_revenue = df.groupby('Product type')['Revenue generated'].sum()

# Sort by total value
sorted_product_types = total_value.sort_values(ascending=False)

# Calculate cumulative sum
cumulative_sum = sorted_product_types.cumsum()

# Calculate cumulative percentage
cumulative_perc = 100 * cumulative_sum / total_value.sum()

# Assign ABC categories
conditions = [
    cumulative_perc <= 70,
    (cumulative_perc > 70) & (cumulative_perc <= 90),
    cumulative_perc > 90
]
choices = ['A', 'B', 'C']

ABC_category = pd.Series(np.select(conditions, choices, default='C'), index=sorted_product_types.index)

sorted_product_types = pd.DataFrame({
    'Total Value': sorted_product_types,
    'Average Defect Rate': average_defect_rate,
    'Total Revenue': total_revenue,
    'ABC Category': ABC_category
})

print(sorted_product_types)
```

EOQ(Economic Order quantity) approach

```python

D = df['Order quantities'].sum()  # annual demand
S = df['Total Cost']  # order cost
H = df['Costs']  # holding cost

# Calculate EOQ
EOQ = np.sqrt((2 * D * S) / H)
rounded_EOQ = np.ceil(EOQ)
print('EOQ:', rounded_EOQ)

```

```python
df_with_EOQ = pd.concat([df, rounded_EOQ.rename('EOQ')], axis=1)
```
